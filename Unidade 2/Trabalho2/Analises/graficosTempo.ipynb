```python
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv("Tabela.csv")
df
```

```python
sns.lineplot(x = "Tamanho", y = "tempo",
             hue = "Algoritmo", style = "Distribuição",
             data = df)
```

```python
df_sem_Bubble = df[df["Algoritmo"] != "Bubble Sort"]
```

```python
sns.lineplot(x = "Tamanho", y = "tempo",
             hue = "Algoritmo", style = "Distribuição",
             data = df_sem_Bubble)
```

```python
df_nLogn = df[df["Algoritmo"].isin(["Merge Sort", "Quick Sort", "STL (Standard Library)"])]
```

```python
sns.lineplot(x = "Tamanho", y = "tempo",
             hue = "Algoritmo", style = "Distribuição",
             data = df_nLogn)
```
