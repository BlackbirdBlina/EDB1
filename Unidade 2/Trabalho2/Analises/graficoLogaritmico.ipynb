```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Tabela.csv")
df
```

```python
fig = sns.lineplot(x = "Tamanho", y = "tempo",
             hue = "Algoritmo", style = "Distribuição",
             data = df)
fig.set(xscale="log")
fig.set(yscale="log")
sns.move_legend(fig, "upper left", bbox_to_anchor=(1, 1))
plt.tight_layout()
fig
```
