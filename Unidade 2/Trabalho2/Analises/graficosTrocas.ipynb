```python
import seaborn as sns
import pandas as pd
```

### Tabela de dados base

```python
df = pd.read_csv("Tabela.csv")
df
```

```python
df = df[df["Algoritmo"] != "STL (Standard Library)"]
```

#### Quantidade de trocas realizadas pelos algoritmos de ordenação em diferentes tamanhos de entrada

```python
sns.lineplot(x = "Tamanho", y = "trocas",
             hue = "Algoritmo", style = "Distribuição",
             data = df)
```

#### Quantidade de trocas realizadas pelos algoritmos de ordenação (sem o Bubble Sort)

```python
df_sem_Bubble = df[df["Algoritmo"] != "Bubble Sort"]
```

```python
sns.lineplot(x = "Tamanho", y = "trocas",
             hue = "Algoritmo", style = "Distribuição",
             data = df_sem_Bubble)
```

#### Algoritmos de ordenação que obtiveram a menor quantidade de trocas realizadas

```python
df_Melhores = df[df["Algoritmo"].isin(["Merge Sort", "Quick Sort", "Selection Sort"])]
```

```python
sns.lineplot(x = "Tamanho", y = "trocas",
             hue = "Algoritmo", style = "Distribuição",
             data = df_Melhores)
```
