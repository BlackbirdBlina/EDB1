```python
import seaborn as sns
import pandas as pd
```

### Tabela de dados base

```python
df = pd.read_csv("Tabela.csv")
df
```

```python
df = df[df["Algoritmo"] != "STL (Standard Library)"]
```

#### Quantidade de comparações realizadas pelos algoritmos de ordenação em diferentes tamanhos de entrada

```python
sns.lineplot(x = "Tamanho", y = "comparações",
             hue = "Algoritmo", style = "Distribuição",
             data = df)
```

#### Quantidade de comparações feitas apenas pelo Bubble Sort

```python
df_Bubble = df[df["Algoritmo"] == "Bubble Sort"]
```

```python
sns.lineplot(x = "Tamanho", y = "comparações",
             hue = "Algoritmo", style = "Distribuição",
             data = df_Bubble)
```

#### Quantidade de comparações feitas apenas pelo Selection Sort

```python
df_Selection = df[df["Algoritmo"] == "Selection Sort"]
```

```python
sns.lineplot(x = "Tamanho", y = "comparações",
             hue = "Algoritmo", style = "Distribuição",
             data = df_Selection)
```

#### Quantidade de comparações feitas apenas pelo Insertion Sort

```python
df_Insertion = df[df["Algoritmo"] == "Insertion Sort"]
```

```python
sns.lineplot(x = "Tamanho", y = "comparações",
             hue = "Algoritmo", style = "Distribuição",
             data = df_Insertion)
```

#### Quantidade de comparações feitas pelo Quick Sort e Merge Sort

```python
df_Melhores = df[df["Algoritmo"].isin(["Merge Sort", "Quick Sort"])]
```

```python
sns.lineplot(x = "Tamanho", y = "comparações",
             hue = "Algoritmo", style = "Distribuição",
             data = df_Melhores)
```
